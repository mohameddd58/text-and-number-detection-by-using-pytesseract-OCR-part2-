```python
import cv2
import os
import pytesseract
```

```python
os.chdir(r"E:")
pytesseract.pytesseract.tesseract_cmd=(r"C:\Program Files\Tesseract-OCR\tesseract.exe")
```

```python
img = cv2.imread("mmm.jpg")
h_img , w_img , _= img.shape
boxes = pytesseract.image_to_boxes(img, lang = "eng" )
print(boxes)
```

```python
for b in boxes.splitlines():
    b=b.split(" ")
    x,y,w,h = int(b[1]),int(b[2]),int(b[3]),int(b[4])
    cv2.rectangle(img, (x,h_img-y), (w,h_img-h), (0,255,0),3)
    cv2.putText(img,b[0],(x,h_img-y+20),cv2.FONT_HERSHEY_PLAIN,2,(0,0,255),2)
```

```python
import matplotlib.pyplot as plt

plt.imshow(img)
```

```python
img = cv2.imread("mmm.jpg")
h_img , w_img , _= img.shape
boxes = pytesseract.image_to_data(img, lang = "eng" )
print(boxes)
```

```python
for x,b in enumerate(boxes.splitlines()):
    if x != 0 :
        b=b.split()
        print(b)
        if len(b)== 12 :
            x,y,w,h = int(b[6]),int(b[7]),int(b[8]),int(b[9])
            cv2.rectangle(img, (x,y), (x+w,y+h), (0,255,0),3)
            cv2.putText(img,b[11],(x,y),cv2.FONT_HERSHEY_PLAIN,2,(0,0,255),2)
```
